## 0. Understanding the signal

We begin the project with a typical proof of concept: How is the signal of a circumbinary planet going to behave? For this we can use the formulas for the radiation of a binary system on a circular trajectory (eq. 4.2 in Maggiore):

$h_+(t)=\frac{4}{r}\left(\frac{GM_c}{c^2}\right)^{5/3}\left(\frac{\pi f_{GW}}{c}\right)^{2/3}\frac{1+\cos^2\theta}{2}\cos{(2\pi \int_0^Tf_{GW}(t')dt'+2\phi)}$ and 

$h_\times(t)=\frac{4}{r}\left(\frac{GM_c}{c^2}\right)^{5/3}\left(\frac{\pi f_{GW}}{c}\right)^{2/3}\cos{\theta}\cos{(2\pi\int_0^Tf_{GW}(t')dt'+2\phi)}$

after this we follow the approach by Cutler (1998) to determine the strain of the GW signal via

$h(t) = \frac{\sqrt{3}}{2}A_{\alpha}
